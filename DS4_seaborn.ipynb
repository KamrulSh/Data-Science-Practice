```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
plt.figure(figsize=(16,6))
sns.distplot(tips['total_bill'], bins=50)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, height=8)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', height=8)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', height=8)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', height=8)
```

```python
sns.pairplot(tips, height=4)
```

```python
sns.pairplot(tips, hue='sex', height=4)
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.countplot( x='day', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
```

```python
sns.boxenplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', split=True, data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', hue='sex', dodge=True, data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', hue='sex', dodge=True, data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
sns.violinplot(x='day', y='total_bill', data=tips, color='white')
```

```python
flight = sns.load_dataset('flights')
```

## Grid

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

```python
dd = sns.PairGrid(iris)
dd.map(plt.scatter)
```

```python
gg = sns.PairGrid(iris, hue='species')
gg.map_diag(sns.distplot)
gg.map_lower(plt.scatter)
gg.map_upper(sns.kdeplot)
```
