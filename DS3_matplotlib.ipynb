```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(1,100,50)
x
```

```python
y = x**2
y
```

```python
plt.figure(figsize=(16,6))
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('X vs Y')
```

```python
plt.figure(figsize=(16,6))
plt.plot(y,x,'r.')
plt.xlabel('Y Label')
plt.ylabel('X Label')
plt.title('Y vs X')
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,3,1)
plt.plot(x,y,'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('X vs Y')

plt.subplot(1,3,2)
plt.plot(y,x)
plt.xlabel('Y Label')
plt.ylabel('X Label')
plt.title('Y vs X')

plt.subplot(1,3,3)
plt.plot(-y,x,'g.')
plt.xlabel('Y Label')
plt.ylabel('X Label')
plt.title('-Y vs X')
```

## object oriented method of matplotlib

```python
fig = plt.figure(figsize=(16,8))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('X vs Y')
```

```python
fig = plt.figure(figsize=(16,8))
plt.xlabel('X Label')
plt.ylabel('Y Label')
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title("LARGER PLOT - (X vs Y)")
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x)
axes2.set_title("SMALLER PLOT - (X vs Y)")
axes2.set_xlim([100,2000])
axes2.set_ylim([10,50])
```

## 2nd method 

```python
fig, axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))
#axes[0].plot(x,y)
plt.tight_layout()
#axes[0].plot(x,y)
for current_axes in axes:
    current_axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))
axes[0].plot(x,y, linestyle='steps')
axes[1].plot(-x,-y, linestyle='--')
axes[2].plot(x,-y*x)
axes[0].set_title("PLOT 1")
axes[1].set_title("PLOT 2")
axes[2].set_title("PLOT 3")
```

```python
fig.savefig("subplots.png", dpi=200)
```

```python
fig = plt.figure(figsize=(15,8))
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y, label="x vs y", linewidth=2, marker='o', markersize=10, markerfacecolor='red')
axes1.plot(x,x*20, label="x vs x*10", linestyle='--')
axes1.legend()
axes1.set_xlim([20,80])
axes1.set_ylim([100,8000])
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
```
